1 Crie um programa que recebe uma lista de números e
- retorne o maior elemento
- retorne a soma dos elementos

- retorne a média dos elementos

```python
lista = [2,2,4,5,6,3,2,2,6,7,8,-3, -2, -10]


def listagem (lista):
    print('Maior numero:',max(lista))
    print('Soma dos elementos:', sum(lista))
    print('Qtd de ocorrencias do primeiro elemento:', lista.count(lista[0]))
    print('Média dos elementos:', (sum(lista))/len(lista))
    
    for item in lista:
        if item == round((sum(lista))/len(lista)):
            valor = item 
    print('Valor mais próximo da média dos elementos:', valor)
    
    neg = 0
    for item in lista:
        if item < 0:
            neg += item 
    print('Soma dos elementos negativos:', neg)
    
    qtd = 0
    for i in range(0, len(lista)-1):
        if (lista[i] == (lista[i+1])):
            qtd+=1
    print('Qtd de vizinhos iguais:', qtd)
    
listagem(lista)
```

2 Faça um programa que receba duas listas e retorne True se são iguais ou False caso
contrario.
Duas listas são iguais se possuem os mesmos valores e na mesma ordem.

```python
lst1 = [1,2,3,4]
lst2 = [1,2,3,4]


def listar(lista1, lista2):
    resp = True
    
    if len(lista1) == len(lista2):
        for i in range(0, len(lista1)):
            if lista1[i] != lista2[i]:
                resp = False
    else:
        resp = False
    return resp

listar(lst1, lst2)
```

3 Faça um programa que receba duas listas e retorne True se têm os mesmos
elementos ou False caso contrário
Duas listas possuem os mesmos elementos quando são compostas pelos mesmos
valores, mas não obrigatoriamente na mesma ordem.

```python
lst1 = [1,2,3]
lst2 = [3,1,2]

    
def listare(lista1, lista2):   
    count = 0
    
    for i in range(0, len(lista1)):
            for u in range(0, len(lista2)):
                if lista1[i] == lista2[u]:
                    count += 1
            
    if (count == len(lista1)):       
        for i in range(0, len(lista2)):
                for u in range(0, len(lista1)):
                    if lista2[i] == lista1[u]:
                        count += 1
    else: 
          return False

        
    if count == len(lista1)+len(lista2):
        return True
    else:
        return False
                    
print(listare(lst1, lst2))
```

8 Escreva um programa para armazenar uma agenda de telefones em um dicionário.
Cada pessoa pode ter um ou mais telefones e a chave do dicionário é o nome da
pessoa. Seu programa deve ter as seguintes funções:
- incluirNovoNome – essa função acrescenta um novo nome na agenda, com um ou
mais telefones. Ela deve receber como argumentos o nome e os telefones.
- incluirTelefone – essa função acrescenta um telefone em um nome existente na
agenda. Caso o nome não exista na agenda, você̂ deve perguntar se a pessoa deseja
inclui-lo. Caso a resposta seja afirmativa, use a função anterior para incluir o novo
nome.
- excluirTelefone – essa função exclui um telefone de uma pessoa que já está na
agenda. Se a pessoa tiver apenas um telefone, ela deve ser excluída da agenda.
- excluirNome – essa função exclui uma pessoa da agenda.
- consultarTelefone – essa função retorna os telefones de uma pessoa na agenda.

```python
agenda = {}

def incluirNovoNome(nome, telefone):
  agenda[nome] = telefone
  print(nome, 'incluido com sucesso.')

def incluirTelefone(nome, telefone):
  if nome in agenda.keys():
    agenda[nome].append(telefone)
    print(agenda[nome], 'incluido com sucesso.')
  else:
    if input("Contato inexistente! Deseja inclui-lo na agenda? (s/n) ") == "s":
      telefones = [telefone]
      incluirNovoNome(nome, telefones)
      print(nome, 'incluido com sucesso.')

def excluirTelefone(nome, telefone):
  if nome in agenda.keys():
    for i in range(0, len(agenda[nome]) - 1):
      if telefone == agenda[nome][i]:
        agenda[nome].pop(i)
        print('Telefone removido com sucesso.')
  else:
    print('Telefone não encontrado.')

def excluirNome(nome):
  if nome in agenda.keys():
    agenda.pop(nome)
    print(nome, 'removido com sucesso.')
  else:
    print(nome, 'não encontrado!')
  
def consultarTelefone(nome):
    if nome in agenda.keys():
      print(agenda[nome])
    else:
      print('Telefone não encontrado.')
```

9 Faça um programa que leia um arquivo texto contendo uma lista de endereços IP e
gere um outro arquivo, contendo um relatório dos endereços IP válidos e inválidos.
O arquivo de entrada possui o seguinte formato:
200.135.80.9
192.168.1.1
8.35.67.74
257.32.4.5
85.345.1.2
1.2.3.4
9.8.234.5
192.168.0.256
O arquivo de saída possui o seguinte formato:
[Endereços válidos:]
200.135.80.9
192.168.1.1
8.35.67.74
1.2.3.4
[Endereços inválidos:]
257.32.4.5
85.345.1.2
9.8.234.5
192.168.0.256

```python
import socket

validos = []
invalidos = []

lista = open("lista.txt")
for ip in lista:
    try:
        socket.inet_aton(ip)
        validos.append(ip)
    except socket.error:
        invalidos.append(ip)
lista.close()

result = open("ipsvalidados.txt", "w+")
result.write("[Endereços válidos:]\n")

for ip in validos:
    result.write(ip)
result.write("[Endereços inválidos:]\n")

for ip in invalidos:
    result.write(ip)
    
result.close()
```

10 A ACME Inc., uma empresa de 500 funcionários, está tendo problemas de espaço
em disco no seu servidor de arquivos. Para tentar resolver este problema, o
Administrador de Rede precisa saber qual o espaço ocupado pelos usuários, e
identificar os usuários com maior espaço ocupado. Através de um programa, baixado
da Internet, ele conseguiu gerar o seguinte arquivo, chamado "usuarios.txt":
alexandre 456123789
anderson 1245698456
antonio 123456456
carlos 91257581
cesar 987458
rosemary 789456125
Neste arquivo, o nome do usuário possui 15 caracteres. A partir deste arquivo, você
deve criar um programa que gere um relatório, chamado "relatório.txt", no seguinte
formato:

ACME Inc. Uso do espaço em disco pelos usuários-----------
Nr. Usuário Espaço utilizado % do uso
1 alexandre 434,99 MB 16,85%
2 anderson 1187,99 MB 46,02%
3 antonio 117,73 MB 4,56%
4 carlos 87,03 MB 3,37%
5 cesar 0,94 MB 0,04%
6 rosemary 752,88 MB 29,16%
Espaço total ocupado: 2581,57 MB
Espaço médio ocupado: 430,26 MB
O arquivo de entrada deve ser lido uma única vez, e os dados armazenados em
memória, caso sejam necessários, de forma a agilizar a execução do programa. A
conversão da espaço ocupado em disco, de bytes para megabytes deverá ser feita
através de uma função separada, que será chamada pelo programa principal. O
cálculo do percentual de uso também deverá ser feito através de uma função, que
será chamada pelo programa principal.

```python
dic = {}
total = 0

def conversor(n):
    result = float(n / (1024*1024))
    return result

def porcent(n, t):
    result = float(n / t) * 100
    return result


usuarios = open("usuarios.txt")

for usuario in usuarios:
    nome, memoria = usuario.split(" ")
    dic[nome] = int(memoria)
    total += int(memoria)

usuarios.close()
media = total / len(dic.keys())

relatorio = open("relatório.txt", "w+")
relatorio.write("ACME Inc.               Uso do espaço em disco pelos usuários.\n")
relatorio.write("--------------------------------------------------------------\n")
relatorio.write("Nr.\tUsuário \tEspaço utilizado\t% do uso\n")

i = 1
for usuario in dic:
    relatorio.write(str(i) + "\t" + usuario + "\t" + "{0:.2f}".format(conversor(dic[usuario])) + " MB\t\t" + "{0:.2f}".format(porcent(dic[usuario], total)) + "%\n")
    i += 1
    

relatorio.write("Espaço total ocupado: " + "{0:.2f}".format(conversor(total)) + "\n")
relatorio.write("Espaço médio ocupado: " + "{0:.2f}".format(conversor(media)))

relatorio.close()
```
